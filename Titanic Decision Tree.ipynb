```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
n_df = df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis ='columns')
n_df.head()
```

```python
inputs = n_df.drop(['Survived'], axis = 'columns')
inputs.head()
```

```python
target = df['Survived']
```

```python
from sklearn.preprocessing import LabelEncoder
le_Pclass = LabelEncoder()
le_Sex = LabelEncoder()
le_Age = LabelEncoder()
le_Fare = LabelEncoder()
```

```python
inputs['n_Pclass'] = le_Pclass.fit_transform(inputs['Pclass'])
inputs['n_Sex'] = le_Pclass.fit_transform(inputs['Sex'])
inputs['n_Age'] = le_Pclass.fit_transform(inputs['Age'])
inputs['n_Fare'] = le_Pclass.fit_transform(inputs['Fare'])
```

```python
inputs
```

```python
n_inputs = inputs.drop(['Pclass', 'Sex', 'Age', 'Fare'], axis = 'columns')
```

```python
n_inputs
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(n_inputs, target)
```

```python
model.score(n_inputs, target)
```

```python
model.predict([[0, 2, 38, 120]])
```
